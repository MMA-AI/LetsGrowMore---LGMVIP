```python
# Import the library
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Get the Image location and the image file name
imagepath = "C:/Users/mmuqt/Downloads/A_chital_deer.jpg"
filename = 'A_chital_deer.jpg'
```

```python
#type of imagepath
type(imagepath)
```

```python
# Read in the image
img = cv2.imread(imagepath)
```

```python
# convert the image to grey scale
grey_filter = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#wait for user to press any keys
#(this is necessary to avoid python kernel form crashing)
cv2.waitKey(0)
```

```python
#Invert the image
invert = cv2.bitwise_not(grey_filter)
```

```python
#Blur the image by gussian function
blur = cv2.GaussianBlur(invert, (21,21),0)
```

```python
#using bitwise_not function to inevrt the image
invertedblur = cv2.bitwise_not(blur)
```

```python
#sketch filter function
sketch_filter = cv2.divide(grey_filter, invertedblur, scale=256.0)
```

```python
#Sketch filter output image
cv2.imwrite("output.png", sketch_filter)
```

```python
#original image and the pencil sketch
invertedblur = 255 - blur
sketch_filter = cv2.divide(grey_filter, invertedblur, scale=256.0)
cv2.imshow("output.png", sketch_filter)
cv2.waitKey(0)
```

```python
# show the image
cv2.imshow("original image", img)
cv2.imshow("sketch_filter", sketch_filter)
cv2.waitKey(0)
```
